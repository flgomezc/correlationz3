```python
%pylab inline
import random
random.seed(0)
```

```python
pos = np.loadtxt('../data/cells/0.cvs', delimiter=',', usecols=[0,1,2])

M=5

Ntotal = 12.651e6
Nbox   = int(Ntotal/(25**3))
Nrand  = M*Nbox
N=pos[:,0].shape[0]



dist = 10**(np.linspace(-2.125,2.125,18)) ## Bins
DIST =logspace(-2,2,17)                   ## Center of Bins
counterDD = np.zeros(17)
counterDR = np.zeros(17)
counterRR = np.zeros(17)

#### Create Random Field
RandomField = np.zeros([M*Nbox,3])
for i in range(M*Nbox):
    for k in range(3):
        RandomField[i,k] = random.random()*10 # Boxlength
```

```python
RandomField.shape
```

```python
fig=figure(figsize=(10,7))
scatter(RandomField[:,0],RandomField[:,1], s=0.1, color='blue')
scatter(pos[:,0],pos[:,1], s=0.1, color='red')
```

```python
#### Counting Pairs in Random Field        
for i in range(Nrand-1):
    for j in range(i+1,Nrand):
        r = 0.0
        for k in range(3):
            r+=(RandomField[i,k]-RandomField[j,k])**2
        r = r**0.5        
        m = 0
        for m in range(counterDD.size):
            if (dist[m]<r) and (r<dist[m+1]):
                counterRR[m] +=1
counterRR = counterRR/( Nrand*(Nrand-1) )
```

```python
scatter(DIST,counterRR, s=20)
yscale('log')
ylim([10**-7,10**1])
xscale('log')
title('Counting Random-Random')
```

```python
### Counting Pairs in Data-Data

for i in range(N-1):
    for j in range(i+1,N):
        r = 0.0
        for k in range(3):
            r+=(pos[i,k]-pos[j,k])**2
        r = r**0.5        
        m = 0
        for m in range(counterDD.size):
            if (dist[m]<r) and (r<dist[m+1]):
                counterDD[m] +=1
counterDD = counterDD*1.0/(N*(N-1))
```

```python
counterDD
```

```python
scatter(DIST,counterDD, s=20)
yscale('log')
ylim([10**-6,10**3])
xscale('log')
title('Counting Data-Data')
```

```python
### Counting pairs Data Random
for i in range(N-1):
    for j in range(0,Nrand):
        r = 0.0
        for k in range(3):
            r+=(pos[i,k]-RandomField[j,k])**2
        r = r**0.5        
        m = 0
        for m in range(counterDR.size):
            if (dist[m]<r) and (r<dist[m+1]):
                counterDR[m] +=1
counterDR = counterDR*1.0/(N*Nrand)
```

```python
scatter(DIST,counterDD, s=20)
yscale('log')
ylim([10**-6,10**1])
xscale('log')
title('Counting Data-Data')
```

```python
CorrFunc = (counterDD - 2*counterDR + counterRR)/ counterRR
```

```python
CorrFunc
```

```python
fig=figure(figsize=[10,6])
scatter(DIST,CorrFunc, s=20)
#yscale('log')
ylim([-10,10])
xscale('log')
title('Correlation Function')
ylabel(r'$\xi(r)$', fontsize=20)
xlabel(r'$r$[Mpc h$^{-1}]$', fontsize=20)
```
