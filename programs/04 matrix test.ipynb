```python
%pylab inline
import random
import timeit
random.seed(0)
```

```python
pos = np.loadtxt('../data/cells/0.cvs', delimiter=',', usecols=[0,1,2])

M=10
Bins = 13

Ntotal = 12.651e6
Nbox   = int(Ntotal/(25**3))
Nrand  = M*Nbox
N=pos[:,0].shape[0]

#Nrand = 4

dist = 10**(np.linspace(-1.125,2.125,Bins+1)) ## Bins
DIST =logspace(-2,2,Bins)                   ## Center of Bins
#counterDD = np.zeros(Bins)
#counterDR = np.zeros(Bins)

#### Create Random Field
RandomField = np.zeros([Nrand,3])
for i in range(Nrand):
    for k in range(3):
        RandomField[i,k] = random.random()*10 # Boxlength
```

```python
start1 = timeit.default_timer()
#### Counting Pairs in Random Field        
counterRR = np.zeros(Bins)

for i in range(Nrand):
    for j in range(0,Nrand):
        r = 0.0
        for k in range(3):
            r+=(RandomField[i,k]-RandomField[j,k])**2
        r = r**0.5        
        m = 0
        for m in range(counterRR.size):
            if ( r >=dist[m] ) and ( r<dist[m+1]):
                counterRR[m] +=1
counterRR = counterRR/( Nrand*(Nrand-1) )
stop1 = timeit.default_timer()
```

```python
print stop1 - start1
```

```python
start2 =  timeit.default_timer()

matrix = np.zeros([Nrand,7])
for k in range(3):
        matrix[:,k]   = RandomField[:,k]

counterRR2 = np.zeros(Bins)
aux = 0

for line in RandomField:
    aux = 0
    for k in range(3):
        matrix[:,k+3] = line[k]
        aux += (matrix[:,k]-matrix[:,k+3])**2
        
    matrix[:,6] = sqrt(aux)
    for m in range(counterRR2.size):
        index = where( ( matrix[:,6] >=dist[m] ) & (matrix[:,6]<dist[m+1]) )[0]
        counterRR2[m] += index.shape
counterRR2 = counterRR2/( 2*Nrand*(Nrand-1) )

stop2 = timeit.default_timer()
```

```python
print stop2 - start2
```
