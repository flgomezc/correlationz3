```python
Conroy = np.genfromtxt("../data/Conroy2006/correlationz3.dat"); 
DIST = np.logspace(-1,2,15)

DD = np.genfromtxt("../results/DD.dat", delimiter=',') 
DR = np.genfromtxt("../results/DR.dat", delimiter=',') 
RR = np.genfromtxt("../results/RR.dat", delimiter=',') 


PCorrFunc = np.zeros(15)
CorrFunc = ( DD -DR + RR)/ (RR)     ## Landy-Szalay

for m in range(15):                 ## Integration over R_parallel
    PCorrFunc[m] = sum(CorrFunc[m])
    
PCorrFunc2 = np.zeros(15)
CorrFunc1 = DD/ RR -1               ## Classic
for m in range(15):                 ## Integration over R_parallel
    PCorrFunc2[m] = sum(CorrFunc[m])
```

Here I plot the correlation function. There is no difference between Landy-Szalay and Peebles estimators.

```python
fig = figure(figsize=[8,6])
title("Projected Correlation Function at z=3 \n Using P-Bolshoi Database", fontsize=24)

plot(DIST,PCorrFunc, label='Landy-Szalay Estimator')
#for m in range(15):
#    plot(DIST,CorrFunc[m])

scatter(DIST,PCorrFunc2, label='Classic Estimator')

plot(Conroy[:,0],Conroy[:,1], label='Conroy et al. 2006')


yscale('log'); xscale('log')
xlabel(r' $r_p [ \mathrm{Mpc\;h}^{-1}]$', fontsize=20)
ylabel(r'$\omega_p(r_p)$', rotation=90, size=20)

xlim(0.1,10)
#ylim(1,500)
legend()
```

The calculated project correlation function is almost 3 times the Conroy function.

```python
fig = figure(figsize=[8,6])
title("Projected Correlation Function at z=3 \n Using P-Bolshoi Database", fontsize=24)

plot(DIST,PCorrFunc*8.0/25, label=r'Landy-Szalay Estimator $\times 8/25$')
#for m in range(15):
#    plot(DIST,CorrFunc[m])

plot(Conroy[:,0],Conroy[:,1], label='Conroy et al. 2006')


yscale('log'); xscale('log')
xlabel(r' $r_p [ \mathrm{Mpc\;h}^{-1}]$', fontsize=20)
ylabel(r'$\omega_p(r_p)$', rotation=90, size=20)

#axhline(y=10)

xlim(0.1,10)
#ylim(1,500)
legend()
```
